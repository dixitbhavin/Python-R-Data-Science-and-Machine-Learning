Vasicek Model : Simulating Interest rate path by vasicek model.

 dr(t)=rate+(eta-gamma x rate) x dt+sqrt_beta x W x sqrt(dt)  ** where W is a Geometic Brownian process

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
rate=0.055
gamma=1
eta=0.08
r_beta=0.015
np.random.seed=1
T=1
days=252
dt=T/days
N=10
```

```python
def Vasicek(rate,gamma,eta,r_beta,dt):
    n=round(T/dt)
    days=252
    t=np.linspace(0,T,n)
    W=[0]+np.random.standard_normal(size=n)
    W=np.cumsum(W)*np.sqrt(dt)
    dr=rate+(eta-gamma*rate)*dt+r_beta*W
    plt.plot(dr)
    return dr
    
    
```

```python
paths=[]
for i in range(0,N-1):
    paths.append(Vasicek(rate,gamma,eta,r_beta,dt))
plt.show()
```
