Generating Random paths using Geometric Brownian motion

St=S0exp((μ−σ^2/2)t−σWt)

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Defining variables

S0=100
sigma=0.2
mu=0.05
T=1
days=252
dt=T/days
N=10
np.random.seed(1)

```

```python
def brownian(S0,mu,sigma,dt,T):
    n=round(T/dt)
    t=np.linspace(0,T,n)
    W=[0]+np.random.standard_normal(size=n)
    W=np.cumsum(W)*np.sqrt(dt)
    X=(mu-0.5*sigma**2)*t+sigma*W
    S=S0*np.exp(X)
    plt.plot(t,S)
    return S

```

```python
paths=[]
for i in range(0,N-1):
    paths.append(brownian(S0,mu,sigma,dt,T))
plt.show()
    
```
